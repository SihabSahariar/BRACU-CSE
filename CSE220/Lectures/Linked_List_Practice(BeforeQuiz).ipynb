```python
class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next
    def prnode(self):
        if self.next == None: #লাস্টের নোড
            print(self.data)
        else:
            print(self.data,'->',end=" ")

class Mylist:
    def __init__(self,lst):
        self.head = None
        tail = None
        
        for i in lst:
            new_node = Node(i,None)
            if self.head == None:
                self.head = new_node
                tail = new_node
            else:
                tail.next = new_node
                tail = new_node
                
                
    def insert_at_first(self,element): #Inseting a node at first
        tmp = self.head
        x = Node(element,tmp)
        self.head = x
                
    
    def insert_at_last(self,element): #inserting a node at last
        new_node = Node(element,None)
        curr = self.head
        while(curr != None):
            xx= curr
            curr = curr.next #লাস্টের অবজেক্ট এর এড্রেস নিলাম 
        #print(xx)
        xx.next = new_node #ঐ ড্রেসের নেক্সট এড্রেস আগে নান ছিল, এখন সেটার সাথে নতুন নোডের লিংক করে দিলাম
    def cnt(self): #counting total nodes
        cnt = 0
        tmp =self.head
        while(tmp):
            cnt+=1
            tmp = tmp.next
        return cnt
    
    def alreadyExist(self,element):
        tmp =self.head
        while(tmp):
            if tmp.data==element:
                return True
            tmp = tmp.next
        return False
                
    def isEmpty(self):
        if self.head == None:
            return True
        else:
            return False
        
    def clear(self):
        if self.isEmpty()==False:
            self.head = None
        else:
            print("Clear can't be done in an Empty List")
        
    def insert_at(self,newElement,index= None): #inserting an element at an index between 0-n 
        if index<0 or index>self.cnt():
            print("Wrong Index")
        elif index==0:
            if self.alreadyExist(newElement) == False:
                self.insert_at_first(newElement)
            else:
                print("The key already exists and does not insert the key.")
                return 
        elif index==self.cnt() or index== None:
            if self.alreadyExist(newElement) == False:
                self.insert_at_last(newElement)
            else:
                print("The key already exists and does not insert the key.")
                return 
        else:
            if self.alreadyExist(newElement) == False:
                c1nt = 1
                tmp =self.head
                while(tmp):
                    if index==c1nt:
                        temp = tmp.next
                        new_node = Node(newElement,temp)
                        tmp.next = new_node
                        break

                    c1nt+=1
                    tmp = tmp.next                   
            else:
                print("The key already exists and does not insert the key.")
                return     
            
    def remove_at(self,index):
        if index<0 or index>self.cnt()-1:
            print("Wrong Index")
        elif index==0:
            tmp = self.head
            self.head = tmp.next
        else:
            ind =1
            tmp =self.head
            while(tmp):
                if ind==index:
                    #print(tmp.data)
                    x= tmp.next.next
                    tmp.next = x
                    break
                ind+=1
                tmp = tmp.next
                    
    def remove(self,deletekey):
        try:
            if self.isEmpty()==False:
                tmp =self.head
                if self.head.data == deletekey:
                    ret = self.head.data
                    self.head = self.head.next
                    return ret
                else:
                    while(tmp):
                        if tmp.next.data==deletekey:
                            ret = tmp.next.data 
                            x= tmp.next.next
                            tmp.next = x
                            return ret
                        tmp = tmp.next
        except:
            print("Doesn't exist the value")
    
    def prnt(self): #print all the nodes
        tmp =self.head
        if tmp == None:
            print("Empty List")
        while(tmp):
            tmp.prnode()
            tmp = tmp.next
            
x = Mylist([3,5,7,9,88,87])
# x.insert_at(55,2)
# x.insert_at(45,2)
# x.insert_at_last(920)
# x.prnt()
x.prnt()
x.clear()
x.prnt()
        
```
