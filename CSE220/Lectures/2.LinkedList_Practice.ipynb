```python
#Add a node after a linked List
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
class MyList:
        def __init__(self):
            self.head = None
            
            
        def addAfterList(self,new_data):
            new_node = Node(new_data)
            
            if self.head == None:
                self.head = new_node
            else:
                last = self.head
                while(last.next):
                    last = last.next
         
            last.next = new_node
            
        def prlist(self):
            temp = self.head
            while(temp):
                print(temp.data)
                temp = temp.next
                
lst = MyList()
el1 = Node(1)
el2 = Node(13)
el3 = Node(78)

lst.head = el1
el1.next = el2
el2.next = el3

lst.addAfterList(90)


lst.prlist()
```

```python
#Take a list as input in a Linkedlist

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class MyList:
    def __init__(self,lst=None):
        self.head = None
        self.tail=None
        if lst:
            for i in lst:
                node = Node(i)
                if self.tail != None:
                    self.tail.next = node
                    self.tail = node
                else:
                    self.head = node
                    self.tail = node

    def prlist(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next    
                
lst = MyList([1,3,7,8,11,78])

lst.prlist()
```

```python
#Take a list as input in a Linkedlist-2

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class MyList:
    def __init__(self,lst=None):
        self.head = None
        if lst:
            for i in lst:
                new_node = Node(i)
                if self.head == None:
                    self.head = new_node
                else:
                    last = self.head
                    while(last.next):
                        last = last.next
                    last.next = new_node
                
    def prlist(self):
        if self.head == None:
            print("Empty List")
            return
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next    
                
lst = MyList([1,45,67,89,66])

lst.prlist()
```

```python
#Linked List | Set 3 (Deleting a node)

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class MyList:
    def __init__(self,lst=None):
        self.head = None
        if lst:
            for i in lst:
                new_node = Node(i)
                if self.head == None:
                    self.head = new_node
                else:
                    last = self.head
                    while(last.next):
                        last = last.next
                    last.next = new_node
                    
                    
    def remove(self, deletekey):
        temp = self.head #storing head node
        
        if temp and temp.data == deletekey:
            self.head = temp.next
            temp = None
            return
        
        while temp: #searching for the node
            if temp.data == deletekey:
                break

            prev = temp #Stroing the temp 
            temp = temp.next
            
        if temp==None: #Not found the node
            return
           
        prev.next = temp.next
        temp = None    
            
    def prlist(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next   
            
            
llist = MyList([7,3,1,2]) 

print ("Created Linked List: ")
llist.prlist() 
llist.remove(22) 
print ("\nLinked List after Deletion of 1:")
llist.prlist() 
```
